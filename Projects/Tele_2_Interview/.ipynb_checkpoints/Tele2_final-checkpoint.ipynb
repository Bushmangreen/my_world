```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import GridSearchCV

#Feature Scaling
from sklearn.preprocessing import StandardScaler

#Classifiers
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
%matplotlib inline
```

# In this task we have to predict the customers that will leave the mobile operator and find the features that affect it most

## 2. Data Cleaning

N1 to N13 are strings so lets convert them to integers

```python
df.loc[:,'N1':'N13'] = df.loc[:, 'N1':'N13'].replace(',','',regex=True)
df.loc[:,'N1':'N13'] = df.loc[:, 'N1':'N13'].astype(float)
```

```python
fig, ax = plt.subplots(figsize=(20, 9))
sns.heatmap(df.loc[:,'N1':'TARGET'].corr(), annot = True)
```

From the correlation heat map we clearly see that N1-N13 columns very strongly correlate with N14-N23 columns


Additionally, N10 and N4, N11 and N5, N20 and N14, N15 and N21 are exactly same since the correlation is 1

In conclusion i can say that we can drop N14-N23 columns since they correlate with N1-N13 columns strongly, more than 0.5 correlation factor.

Also, we can drop N10 and N11 columns since they are copies of N4 and N5

## 4. Building Machine Learning Models

I will explore different Classification Models to find the best fitting one:
- DecisionTree
- GaussianNB
- KNeighbors
- LogisticRegression
- LinearDiscriminant
- SVM
- Random Forest Classifier

For the features i will use only ['C2','C3','N1','N2','N3','N4','N5','N6','N7','N12','N13'] columns since rest of them strongly correlate with those above

Since we are not given with the test data we gonna divide train.csv into train(70%) and test(30%)

```python
X = df.loc[:,['C2','C3','N1','N2','N3','N4','N5','N6','N7','N12','N13']]
y = df.TARGET
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Let do training and fitting, after that i will use confusion matrix to choose the best model

```python
# DecisionTree
print('DecisionTree')
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
print(classification_report(y_test, dt_predictions))
print(confusion_matrix(y_test, dt_predictions))

# GaussianNB
print('GaussianNB')
gauss_model = GaussianNB()
gauss_model.fit(X_train, y_train)
gauss_predictions = gauss_model.predict(X_test)
print(classification_report(y_test, gauss_predictions))
print(confusion_matrix(y_test, gauss_predictions))

# KNeighbors
print('KNeighbors')
kn_model = KNeighborsClassifier()
kn_model.fit(X_train, y_train)
kn_predictions = kn_model.predict(X_test)
print(classification_report(y_test, kn_predictions))
print(confusion_matrix(y_test, kn_predictions))

# LogisticRegression
print('LogisticRegression')
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
print(classification_report(y_test, lr_predictions))
print(confusion_matrix(y_test, lr_predictions))

# LinearDiscriminant
print('LinearDiscriminant')
ld_model = LinearDiscriminantAnalysis()
ld_model.fit(X_train, y_train)
ld_predictions = ld_model.predict(X_test)
print(classification_report(y_test, ld_predictions))
print(confusion_matrix(y_test, ld_predictions))


# RFS
print('RFS')
rfs_model = RandomForestClassifier()
rfs_model.fit(X_train, y_train)
rfs_predictions = rfs_model.predict(X_test)
print(classification_report(y_test, rfs_predictions))
print(confusion_matrix(y_test, rfs_predictions))
```

So from the results above, the best result so far is Random Forest Classifier. It Labeled correctly 28% of those who wanted to swithch mobile operator (ones), additionally it labeled 96% of those who did not change (zeros). Overall, average score is 94%.

We can improve the result little bit by tuning hyoer parameters:

```python
# RFS
rfs_model = RandomForestClassifier(min_samples_leaf=10, n_estimators=10)
rfs_model.fit(X_train, y_train)
rfs_predictions = rfs_model.predict(X_test)
print(classification_report(y_test, rfs_predictions))
print(confusion_matrix(y_test, rfs_predictions))
```

Lets ses what will happen to the results if we used all the Features N1->N23

```python
X = df.loc[:,'C2':'N23']
y = df.TARGET
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
# RFS
rfs_model = RandomForestClassifier(min_samples_leaf=10, n_estimators=10)
rfs_model.fit(X_train, y_train)
rfs_predictions = rfs_model.predict(X_test)
print(classification_report(y_test, rfs_predictions))
print(confusion_matrix(y_test, rfs_predictions))
```

So we can see that, even if we dropped the columns that have strong correlation, accuracy score did not change
